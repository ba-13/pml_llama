```python
import matplotlib.pyplot as plt
import pickle as pk
import numpy as np
```

```python
mc_data = []
with open("./tmp/test_accuracies_mc2.pickle", "rb") as f:
    mc_data = pk.load(f)
ppd_data = []
with open("./tmp/test_accuracies_ppd2.pickle", "rb") as f:
    ppd_data = pk.load(f)
with open("./tmp/test_accuracies_random2.pickle", "rb") as f:
    random_data = pk.load(f)

ppd_data = ppd_data[:24]
mc_data = mc_data[:24]
random_data = random_data[:24]
```

```python
len(mc_data), len(ppd_data), len(random_data)
```

```python
plt.figure(figsize=(12, 8))
x = np.arange(200, 100 * len(ppd_data) + 200, 100)
plt.plot(x, ppd_data, label="ppd_llama acquisition", marker="x")
plt.plot(x, mc_data, label="monte carlo acquisition", marker="o")
plt.plot(x, random_data, label="random acquisition", marker="*")
plt.xlabel("training dataset size")
plt.ylabel("accuracy")
plt.legend()
plt.grid()
```
